####  IV sweep Timescan monitor

Monitor plot updates automatically at a user-specified interval.

RLH 1/1/2022

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd
```

```python
%matplotlib notebook
```

```python
def animate_timescan(i, filename):
    ts_df = pd.read_csv(filename)
    time = ts_df["Elapsed Time"]
    efficiency = ts_df["PCE(%)"]
    ax1.clear()
    ax1.plot(time/60.0, efficiency, 'o')
    plt.xlabel("Time (hours)")
    plt.ylabel("PCE (%)")
```

```python
path = r"C:\Users\LabUser\Documents\randy\Data_032022"
filename = "timeseries_parameters_010222"
```

```python
# Run this code to start the animation
# Make sure you have the correct  path and filename!
update_interval = 5000  # units of ms.

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

ani = animation.FuncAnimation(fig, animate_timescan, interval=update_interval, fargs = [path + "\\" + filename])
plt.show()
```
